```python
import requests
from google.colab import userdata   # ‚Üê this is the key import

# Get token securely from Colab Secrets
HF_TOKEN = userdata.get('HF_TOKEN')

# Optional: quick check (remove in final version)
if not HF_TOKEN:
    raise ValueError("HF_TOKEN not found in Colab secrets. Please add it!")

API_URL = "https://router.huggingface.co/v1/chat/completions"

headers = {
    "Authorization": f"Bearer {HF_TOKEN}",
    "Content-Type": "application/json"
}

payload = {
    "model": "meta-llama/Llama-3.2-3B-Instruct",
    "messages": [
        {"role": "user", "content": "The capital of Japan is"}
    ],
    "max_tokens": 64,
    "temperature": 0.7,
    "stream": False
}

response = requests.post(API_URL, headers=headers, json=payload)

if response.status_code == 200:
    result = response.json()
    print(result["choices"][0]["message"]["content"])
else:
    print("Error:", response.status_code, response.text)
```
